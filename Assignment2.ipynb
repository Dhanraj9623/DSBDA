# Handling missing values

```python
df.ffill(inplace=True)
df.isna().sum()
```

```python
df
```

# 2. Outliers

```python
sns.boxplot(df)
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1 
ub = q3 + 1.5 * iqr
lb = q1 - 1.5 * iqr
```

```python
ub,lb
```

```python
filtered_data = df[~((df > ub) | (df < lb))]
```

```python
filtered_data.dropna(inplace = True)
filtered_data
```

# outliers using another method z-score

```python
zscore = (df - df.mean())/df.std()
newdf = df[(zscore<3) & (zscore> -3)]
newdf
```
